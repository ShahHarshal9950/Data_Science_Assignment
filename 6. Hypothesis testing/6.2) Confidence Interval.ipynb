```python
import numpy as np
import math
from scipy.stats import t,norm
```

```python
n = 15
Data = 1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29 ####(in millions of characters)
```

```python
## (A) 99% Confidence Interval Using Sample Standard Deviation
```

```python
## Use T-Test because sample size is small n<30 , population standard daviation is unknown thats why use T-Test
```

```python
n = len(Data)
sample_mean = np.mean(Data)
sample_std = np.std(Data, ddof=1)
alpha = 0.01

print("Sample Size:" ,n)
print("Sample Mean:", round(sample_mean,4))
print("Sample standard Daviation:", round(sample_std,4))
```

```python
## Degree of freedom
df = n - 1

## Critical value
t_critical = t.ppf(1-alpha/2 , df)

## margin of errors
margin = t_critical * (sample_std/math.sqrt(n))

## Confidence Interval
confi = (sample_mean - margin, sample_mean + margin)

print("\n99% Confidence Interval using t-distribution:")
print("Lower bound:", round(confi[0],4))
print("Upper bound:", round(confi[1],4))
```

```python
## (B). 99% Confidence Interval using Z-distribution
## use z-Test because population standard deviation is known.
```

```python
sigma = 0.2
```

```python
## Critical value
z_critial = norm.ppf(1 - alpha / 2)

## margin of errors
margin_z = z_critial * (sigma / math.sqrt(n))

## Confidence Interval
CI = (sample_mean - margin_z , sample_mean + margin_z)

print("\n99% Confidence Interval using Z-distribution:")
print("Lower bound:", round(CI[0],4))
print("Upper bound:", round(CI[1],4))
```
