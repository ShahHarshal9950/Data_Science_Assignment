```python
import math
from scipy.stats import norm
```

```python
sample_mean = 3050
X = 600
MU = 1000 + 5 * X 
SD = 25
sigma = 5* 25
n = 25           #Sample Size
Alpha = 0.05
```

```python
# Hypotheses statement:
## Null Hypothesis(H0) : The mean weekly operating cost follows the theoretical model 
## Alternative Hypothesis(H1) : The mean weekly operating cost is higher than the model 
```

```python
# 2. Z_Test Calculation
# n = 25, population is normally distributed and sigma is known,
# therefore Z-test is valid even for small samples
```

```python
z = (sample_mean - MU) / (sigma / math.sqrt(n))
print("Z_Value:", z)
```

```python
# 3. Determine the Probability 
```

```python
z_critical = norm.ppf(1 - alpha)
print("z_critical:", z_critical)
```

```python
print("Theoretical Mean Cost:", MU)
print("Z-test Statistic:", z)
print("Critical Z-value:", z_critical)
```

```python
## Decision
if z > z_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
P_value = 1 - norm.cdf(z)
P_value
```

```python
## Conclusion
# Based on the Z-test results, the calculated Z-value (−38.0) is far less than the critical Z-value (1.645) at the 5% significance level.
# Therefore, we fail to reject the null hypothesis.
# This indicates that there is no strong statistical evidence to support the restaurant owners’ claim that the mean weekly operating costs are higher
# than what the theoretical cost model suggests. In fact, the observed sample mean weekly cost is substantially lower than the model’s predicted cost.
```
