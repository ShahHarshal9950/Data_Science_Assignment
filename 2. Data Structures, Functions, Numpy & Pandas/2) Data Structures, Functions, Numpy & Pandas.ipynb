### Section A: Data Structures & Control Structures

```python
# 1.	Create a list of 5 integers. Perform and print the result of the following operations: 
# append, extend, insert, remove, pop, clear, index, count, sort, and reverse.
Data = [10,20,30,40,50]
Data.append(60)
print("List after appending:", Data)
Data.extend([11,12])
print("List after extending:", Data)
Data.insert(1,300)
print("List after inserting:", Data)
Data.remove(300)
print("List after removing:", Data)
Data.pop(7)
print("List after poping:", Data)
print("index of 11:", Data.index(11))
print("count of 11:", Data.count(11))
Data.sort()
print("List of Shorting:", Data)
Data.reverse()
print("List of Reversing:", Data)
Data.clear()
print("Clean the list:", Data)
```

```python
# 2. Create a tuple that stores 3 student names. Try changing the second name in the tuple. What happens? Explain why.
students = ("Amit", "Neha", "Rahul")
print(students)
students[1]="Priya"

# Explanation:
# This happens because tuples are immutable in Python.
# Immutability means that once a tuple is created, its elements cannot be changed, added, or removed.
```

```python
# 3. Create a set of integers with some duplicate values. Print the set and explain the output.
numbers = {1, 2, 3, 4, 2, 3, 5}
print(numbers)

# Explanation:
# A set in Python automatically removes duplicate values.
# Even though 2 and 3 were added more than once, they appear only once in the output.
# This happens because sets store unique elements only and do not allow duplicates.
```

```python
# 4. Create a dictionary with the keys: 'name', 'age', and 'city'. Update the age and add a new key 'email'. Print the final dictionary.
Person = {
    "Name" : "Harshal",
    "Age" : 24,
    "City" : "Banglore"
}

Person["Age"] = 26
Person["Email"] = "harshal2@gmail.com"
print(Person)
```

```python
# 5. Write a script that checks if a person is eligible to vote (age ≥ 18). Take age as a variable and print the appropriate message.

age = int(input("Put the age:"))

if age >= 18:
    print("The person is eligible to vote.")
else:
    print("The person is not eligible to vote.")
```

```python
# 6. Given a 'marks' variable, print the grade:
# 90 and above: 'A'
# 75–89: 'B'
# 50–74: 'C'
# Below 50: 'Fail'

marks =int(input("Enter the marks:" ))
if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 50:
    grade = "C"
else:
    grade = "Fail"

print("Grade:", grade)
```

```python
# 7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

number = int(input("Give a number:"))

if number >= 0 and number % 2 == 0:
    num = "Number is positive & even"
elif number >= 0 and number % 2 != 0:
    num = "Number is positive but odd"
elif number < 0 and number % 2 == 0:
    num = "Number is nagative & even"
elif number < 0 and number % 2 != 0:
    num = "Number is nagative but odd"
elif number == 0:
    num = "Number is zero"

print("Number:", num)
```
